```python
import scikitplot as splot
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics import roc_curve, auc, roc_auc_score
from savePlot import savePlot
```

```python
df = pd.read_csv('./pred.txt', sep=' ')

outcyte = df['OutCyte'].values
secp = df['SecretomeP'].values
print outcyte.shape, secp.shape
print outcyte
```

```python
y_data = np.concatenate((np.ones(18), np.zeros(20)))
y_pred = np.vstack((outcyte, secp)).T
print y_pred.shape
print roc_auc_score(y_data, outcyte), auc(y_data, outcyte)
```

```python
print y_pred[:, 1]
```

```python
def plot_roc(y_data, y_score):
    fpr, tpr, _ = roc_curve(y_data, y_score[:, 0])
    fpr_s, tpr_s, _ = roc_curve(y_data, y_score[:, 1])

    roc_auc = auc(fpr, tpr)
    roc_auc_s = auc(fpr_s, tpr_s)

    plt.figure(figsize=(12, 10))
    lw = 2.5
    plt.plot(fpr, tpr, lw=lw, label='OutCyte (area=%.2f)'%roc_auc)
    plt.plot(fpr_s, tpr_s, lw=lw, label='SecretomeP (area=%.2f)'%roc_auc_s, linestyle=':')

    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=20)
    plt.ylabel('True Positive Rate', fontsize=20)
    plt.title('ROC curve for independent dataset', fontsize=20)
    plt.legend(loc='lower right', fontsize=16)
    plt.savefig('roc.pdf', dpi=300)
    #savePlot('ROC_curve')
    plt.show()
```

```python
print y_data.shape
print outcyte.shape
plot_roc(y_data, y_pred)
```
